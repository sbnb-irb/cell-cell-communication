# Supplementary figure 4: 2D representation of GEx profiles (UMAP)

In this notebook, we explore if we are able to split the spatial regions using a unsupersived dimensionality reduction approach

```python
global_path = './cell-cell-communication/' ## Path to the github downloaded repository
```

```python
import scanpy as sc
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import numpy as np
from collections import Counter
from matplotlib.lines import Line2D
import sys
sys.path.insert(1, global_path + 'Functions/')
from downstream_analysis import obtain_adata_embeddings 

title_size = 12
labelout_size = 10
label_size = 8
ticks_size = 6
```

```python
fig, axs = plt.subplots(figsize=(3,3), dpi=200, facecolor='white')

adata_file = global_path + 'Data/Brain/Input_CP/adata_L5_region_ass.h5ad'
adata = obtain_adata_embeddings(adata_file, ['CC1', 'CC2'])


axs = sns.scatterplot(x=adata.obsm['X_umap'][:, 0], y=adata.obsm['X_umap'][:, 1], hue=adata.obs['celltype_loc'], palette=['#B16BA8', '#F39426'], s=15, ax = axs)
axs.set_xticks([])
axs.set_yticks([])
axs.set_title('L5 IT cells UMAP projection', fontsize=title_size)

handles, labels = axs.get_legend_handles_labels()
new_labels = ['Outer', 'Inner']
legend = axs.legend(handles, new_labels, frameon = False)
```

In this case, UMAP projections of scRNA-seq data successfully separated the two groups, suggesting spatial gene expression differences when the 2000 most variable genes are visualized in a two-dimensional framework
